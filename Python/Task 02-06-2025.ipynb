```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('customers.csv')
```

```python
df
```

```python
gender = df['Gender'].value_counts()
```

```python
plt.figure(figsize =(12, 5))
gender.plot(kind='bar')
plt.title('Gender Info')
plt.xlabel('Gender')
plt.ylabel('Gender Count')
plt.show()
```

```python
# Group by Gender and calculate average spending score
spending_by_gender = df.groupby('Gender')['SpendingScore'].mean()

print(spending_by_gender)
```

```python
# Create scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['SpendingScore'], alpha=0.6, color='teal')
plt.title('Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

```python
correlation = df['Age'].corr(df['SpendingScore'])
print("Correlation between Age and Spending Score:", correlation)
```

```python
# Group by Category and calculate average spending
avg_spending = df.groupby('PurchaseCategory')['AnnualIncome'].mean()

# Sort by highest average
avg_spending = avg_spending.sort_values(ascending=False)

print(avg_spending)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['SpendingScore'], df['AnnualIncome'], color='green', alpha=0.6)
plt.title('Income vs Spending Amount')
plt.xlabel('Income')
plt.ylabel('AnnualIncome')
plt.grid(True)
plt.show()

# Optional: correlation check
correlation = df['SpendingScore'].corr(df['SpendingScore'])
print("Correlation between Income and Spending:", correlation)
```

```python
# Count of each category
category_counts = df['PurchaseCategory'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Purchase Category Distribution')
plt.ylabel('')  # Hide y-label
plt.show()
```

##  Final Summary & Marketing Insights

###  Who are the High Spenders?
- High spenders are individuals with **higher Spending Scores** or **Spending Amounts**.
- They are typically in the **younger age group (1835)**, based on analysis.
- Often belong to **mid-to-high income brackets**, depending on the correlation with income.

###  Patterns by Gender or Age
- **Gender**: Females tend to have **higher average spending scores** compared to males.
- **Age**: Younger customers (especially those under 35) are more likely to spend more.
- **Category Preference**:
  - Younger females often spend more on **Clothing**.
  - Males may spend more on **Electronics**, but less frequently.
  - **Groceries** have lower average spending but higher frequency across all groups.

###  Marketing Recommendations
- Focus promotions and campaigns on **younger demographics (1835)**.
- Use **targeted ads for females** in fashion/clothing categories.
- Promote **bundle offers or tech deals** for males in electronics.
- Introduce **loyalty programs** for grocery buyers to boost retention.
- Use **personalized offers** based on age, income, and past purchases to improve ROI.
